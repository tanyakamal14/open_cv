```python
# Step 4: Convert to Grayscale
gray =cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)# Step 5: Resize the Image to 300×300
a=[image,gray]
resized = cv2.resize(image,(300, 300))
cv2_imshow(resized)
```
